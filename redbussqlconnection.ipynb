```python
import re

#Strip whitespace from all string columns
final_df = final_df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
```

```python
# 2. Convert 'Price' column to numeric (remove 'INR ' and convert to int)
final_df['Price'] = final_df['Price'].str.replace('INR ', '', regex=True).astype(float)
final_df['Price'] = final_df['Price'].round().astype(float) 

print(final_df)
```

```python
#Extract numeric values from 'Seats_Available' (remove 'Seats available')
final_df['Seats_Available'] = final_df['Seats_Available'].str.extract('(\d+)').astype(float)
```

```python
#Extract rating from 'Ratings' column (ignore review count)
final_df['Ratings'] = final_df['Ratings'].str.extract(r'(\d+\.\d+)').astype(float)
```

```python
# 5. Convert 'Total_duration' to total minutes
def duration_to_minutes(duration):
    match = re.search(r'(\d+)h (\d+)m', duration)  # Format: '27h 40m'
    if match:
        return int(match.group(1)) * 60 + int(match.group(2))
    elif 'h' in duration:  # If only hours are present
        return int(re.search(r'(\d+)h', duration).group(1)) * 60
    elif 'm' in duration:  # If only minutes are present
        return int(re.search(r'(\d+)m', duration).group(1))
    return np.nan  # If format is incorrect

final_df['Total_duration'] = final_df['Total_duration'].apply(duration_to_minutes).astype(float)
```

```python
# 6. Handle missing values
final_df.fillna({'Ratings': final_df['Ratings'].median()}, inplace=True)  # Fill missing ratings with median
final_df.dropna(inplace=True)  # Drop rows with remaining NaN values

# Display cleaned data
print(final_df.info())
print(final_df.head())
```

```python
import mysql.connector
import pandas as pd

# MySQL Connection
host = "localhost"        
user = "root"      
password = "#Vanathi13" 
database = "bus_routes"     

# Connect to MySQL Server (Without specifying database first)
conn = mysql.connector.connect(host=host, user=user, password=password)
cursor = conn.cursor()

#Create Database if not exists
cursor.execute(f"CREATE DATABASE IF NOT EXISTS {database}")
conn.commit()

#Connect to the newly created database
conn.database = database
```

```python
#Create Table
table_creation_query = """
CREATE TABLE IF NOT EXISTS bus_details (
    Bus_name VARCHAR(255),
    Bus_type VARCHAR(255),
    Start_time VARCHAR(10),
    End_time VARCHAR(10),
    Total_duration FLOAT,
    Price FLOAT,
    Seats_Available FLOAT,
    Ratings FLOAT,
    Route_link TEXT,
    Route_name VARCHAR(255)
);
"""
cursor.execute(table_creation_query)
conn.commit()

# Insert Data into MySQL Table
insert_query = """
INSERT INTO bus_details (Bus_name, Bus_type, Start_time, End_time, Total_duration, Price, Seats_Available, Ratings, Route_link, Route_name) 
VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s);
"""

# Convert DataFrame to List of Tuples
data_to_insert = [tuple(row) for row in final_df.itertuples(index=False, name=None)]

#Insert in Batches for Efficiency
batch_size = 1000  # Adjust batch size if needed
for i in range(0, len(data_to_insert), batch_size):
    cursor.executemany(insert_query, data_to_insert[i:i + batch_size])
    conn.commit()

#Close Connection
cursor.close()
conn.close()

print("Data successfully inserted into MySQL under 'bus_routes' schema")
```
