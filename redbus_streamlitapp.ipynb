```python
import streamlit as st
import mysql.connector
import pandas as pd
```

```python
# Function to get all bus data from MySQL
def get_data():
    connection = mysql.connector.connect(
        host='localhost',
        user='root',
        password='#Vanathi13', 
        database='bus_routes')
    query = "SELECT * FROM bus_details"
    df = pd.read_sql(query, connection)
    connection.close()
    return df

# Function to get data from MySQL with filters
def get_filtered_data(bus_type_filter, route_name, price_range, rating, availability):
    # MySQL connection
    connection = mysql.connector.connect(
        host='localhost',
        user='root',
        password='#Vanathi13',
        database='bus_routes')

    # SQL query
    query = "SELECT * FROM bus_details WHERE 1=1"

    # Applying bus type filters dynamically
    if bus_type_filter:
        conditions = []

        if 'AC' in bus_type_filter:
            conditions.append("Bus_type REGEXP 'AC|A/C|ac' AND Bus_type NOT REGEXP 'Non AC|NON A/C|NON-AC|NON-A/C|non AC'")

        if 'Non AC' in bus_type_filter:
            conditions.append("Bus_type REGEXP 'Non AC|NON A/C|NON-AC|NON-A/C|non AC'")

        if 'Sleeper' in bus_type_filter:
            conditions.append("Bus_type REGEXP 'Sleeper|sleeper'")

        if 'Semi Sleeper' in bus_type_filter:
            conditions.append("Bus_type REGEXP 'Semi Sleeper|semi sleeper'")

        if 'Seater' in bus_type_filter:
            conditions.append("Bus_type REGEXP 'Seater|seater'")

        # Join conditions with OR
        query += " AND (" + " OR ".join(conditions) + ")"

    # Apply other filters
    if route_name:
        query += f" AND Route_name = '{route_name}'"
    if price_range:
        query += f" AND Price BETWEEN {price_range[0]} AND {price_range[1]}"
    if rating is not None:
        query += f" AND Ratings >= {rating}"
    if availability:
        query += f" AND Seats_Available >= {availability[0]}"

    # Fetch the filtered data
    df = pd.read_sql(query, connection)
    connection.close()
    return df
```

```python
# Load initial data from MySQL
df = get_data()

# Streamlit UI
st.title("Bus Routes Information")
st.write("This application allows you to filter bus routes based on various criteria.")

# Bus Type Filter (Multiple selection for dynamic filtering)
bus_types = ['AC', 'Non AC', 'Sleeper', 'Semi Sleeper', 'Seater']
bus_type_filter = st.multiselect(
    'Select Bus Type',
    bus_types,
    help="Choose the bus type(s) you want to filter by.")

# Route Filter
route_filter = st.selectbox('Select Route', [''] + list(df['Route_name'].unique()))

# Price Range Filter
min_price = df['Price'].min()
max_price = df['Price'].max()
price_range = st.slider('Select Price Range', min_value=int(min_price), max_value=int(max_price), value=(int(min_price), int(max_price)))

# Star Rating Filter
rating_filter = st.slider('Select Minimum Rating', 0.0, 5.0, 0.0)

# Seats Available Filter
min_seats = df['Seats_Available'].min()
max_seats = df['Seats_Available'].max()
availability_filter = st.slider('Select Minimum Seats Available', min_value=int(min_seats), max_value=int(max_seats), value=(int(min_seats), int(max_seats)))

# Get filtered data based on selected filters
filtered_data = get_filtered_data(bus_type_filter, route_filter, price_range, rating_filter, availability_filter)

# Display filtered data
st.write(f"{len(filtered_data)} buses found!")
st.dataframe(filtered_data)
```
